```python
import numpy as np
import scipy as sp
from plotly.subplots import make_subplots
np.set_printoptions(precision=1, linewidth=100)
```

```python
r = 0.5015
```

```python
T=np.full(50, 20.)
T[0]=40
T[-1]=80.
fig=make_subplots()
fig.add_scatter(x=np.arange(T.size), y=T, mode='lines')
for i in range(1500):
    T[1:-1] = T[1:-1] + r*(T[2:] - 2*T[1:-1] + T[:-2])
    if i % 100 == 0:
        fig.add_scatter(x=np.arange(T.size), y=T, mode='lines')
fig.update_layout(width=600, height=400, template='plotly_dark', showlegend=False)
```

```python
Tleft = 40.
Tright= 80.
M=20
Tinitial = np.full(M, 20.)

r= 2
Adiags = np.vstack([np.r_[0, np.full(M-1,-r)],
                    np.full(M,1+2*r),
                    np.r_[np.full(M-1,-r), 0 ]])
```

```python
Thistory=[Tinitial]
Told = Tinitial.copy()
for i in range(100):
    tempT =np.r_[Tleft, Told, Tright]
    b = r*tempT[2:] + (1-2*r)*tempT[1:-1] + r*tempT[:-2]
    b[0]+=r*Tleft
    b[-1]+=r*Tright
    Tnew = sp.linalg.solve_banded((1,1), Adiags, b)
    Thistory.append(Tnew)
    Told=Tnew.copy()

T=np.c_[np.full(len(Thistory), Tleft),np.vstack(Thistory),np.full(len(Thistory),Tright)]
```

```python
fig2=make_subplots()
xplot = np.arange(T.shape[1])
for i in range(T.shape[0]):
    if i % 10 == 0:
        fig2.add_scatter(x=xplot, y=T[i,:], mode='lines')
fig2.update_layout(width=600, height=400, template='plotly_dark', showlegend=False)
```
