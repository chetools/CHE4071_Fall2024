```python
from scipy.integrate import solve_ivp
import numpy as np

from plotly.subplots import make_subplots
```

```python
q = 100 #L/min
c_ai = 1 #mol/L
Ti = 350 #K
V = 100 #L
rho = 1000 #g/L
C = 0.239 #J/g K
neg_Hr = 5e4 #J/mol
ER = 8750 #K
k0 = 7.2e10 #/min
UA = 5e4 #J/min K
Tc0 = 300. #K
c_a0 = 0.5#mol/L
T0 = 350 #K
```

```python
def rhs(t,vec):
    T, c = vec

    k = k0*np.exp(-ER/T)
    dca = q/V*(c_ai - c) - k*c
    dT = q/V*(Ti - T) + k*c * (neg_Hr)/(rho*C) - UA* (T - 305.) / (rho * V * C)
    return  dT, dca
```

```python
tend = 10. #min
res = solve_ivp(rhs, (0,tend), [T0,c_a0], method="Radau", dense_output=True)

tplot = np.linspace(0,tend,500)
Tplot , cplot = res.sol(tplot)
```

```python
fig = make_subplots(1,2)
fig.update_layout(width=600, height=600)
fig.add_scatter(x=tplot, y=Tplot, row=1, col=1, name='T')
fig.add_scatter(x=tplot, y=cplot, row=1, col=2, name='Ca')
fig.update_layout(height=400,width=800, template='plotly_dark')
```
