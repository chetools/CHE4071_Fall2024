```python
from scipy.integrate import solve_ivp
import numpy as np
from scipy.special import expit
from plotly.subplots import make_subplots
```

```python
N=1000
q = 1 #L/min
total_V = 1 #L
V = total_V/N  #L volume per stirred tank
ca0 = 0.2
```

```python
def spike(t, t1, t2, a, s):
    return a*(expit(s*(t-t1)) - expit(s*(t-t2)))
```

```python
tplot = np.linspace(0,10,1000)
y=spike(tplot, 2., 2.5, 3., 100.)
```

```python
fig=make_subplots()
fig.add_scatter(x=tplot, y=y)
fig.update_layout(height=400, width=600, template='plotly_dark')
```

```python

def rhs(t,c):

    cin = spike(t, 0.1, 0.15, 3., 100.) + ca0
    c_ins = np.r_[cin, c[:-1]]
    dca = q/V*(c_ins - c)
    return dca

```

```python
tend = 1.
res = solve_ivp(rhs, (0,tend), np.r_[np.repeat(ca0,N)], method="Radau", dense_output=True)

tplot = np.linspace(0,tend,11)

cs = res.sol(tplot).T
x= np.arange(N)
fig=make_subplots()
for c in cs:
    fig.add_scatter(x=x, y=c)
fig.update_layout(height=400, width=600, template='plotly_dark', showlegend=True)
```
